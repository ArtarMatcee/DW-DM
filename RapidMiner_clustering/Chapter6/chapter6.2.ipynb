```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
DataPath = "data/Chapter06Exercise.csv"
Data = pd.read_csv(DataPath)
```

```python
Data.head()
```

# Clean attribute

```python
cleanData = Data.drop(['Student_ID', 'First_Name', 'Last_Name', 'Expulsion', 'Suspension', 'Gender', 'Grade'], axis=1)
cleanData.head()
```

# Matrix scatter plot to find correlation data

```python
pd.plotting.scatter_matrix(cleanData, alpha=0.2, figsize=(10, 10), diagonal='kde')
plt.show()
```

# K-means & Plot data

```python
# Create a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=3, n_init='auto', max_iter=100, random_state=1)

# Fit the data
kmeans.fit(cleanData)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Get the cluster labels
labels = kmeans.labels_

# Plot the data
Tardies = cleanData['Tardies']
Absences = cleanData['Absences']


plt.scatter(Tardies, Absences, c=labels, cmap="rainbow")


plt.xlabel('Tardies')
plt.ylabel('Absences')

plt.show()
```
